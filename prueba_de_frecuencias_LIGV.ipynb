```python
import numpy as np
import scipy.stats as stats


datos = [
    0.78961, 0.05230, 0.10699, 0.55877, 0.14151, 0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989, 0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121, 0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850, 0.79882, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264, 0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272, 0.57410, 0.26593, 0.85903, 0.43308, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187, 0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954, 0.58244, 0.85853, 0.86752, 0.38729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49883, 0.75103, 0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724, 0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

def prueba_frecuencias(datos, k=10):
    n = len(datos)
    # 1. Determinar las frecuencias observadas en k intervalos
    # np.histogram cuenta cuántos datos caen en cada bin
    observados, bins = np.histogram(datos, bins=np.linspace(0, 1, k + 1))

    #nivel de significancia
    alpha = 0.05



    # frecuencia esperada
    fe = n / k

    # estadistico de prueba con fo como frecuencia observada y fe como frecuencia esperada
    ji_calc = sum([(fo - fe)**2 / fe for fo in observados])

    # obtener el valor crítico al mismo tiempo que obtenemos el grado de libertad
    gl = k - 1
    valor_critico = stats.chi2.ppf(1 - alpha, gl)

    # creamos la tabla como  la del video
    print(f"--- Prueba de Frecuencias (n={n}) ---")
    print(f"{'Intervalo':<15} | {'Obs (Oi)':<8}")

    for i in range(k):
        intervalo = f"[{bins[i]:.1f} - {bins[i+1]:.1f})"
        print(f"{intervalo:<15} | {observados[i]:<8}")


    print(f"\n\nji-cuadrado calculado: {ji_calc:.4f}")
    print(f"ji-cuadrado crítico:   {valor_critico:.4f} \ngrados de libertad:   {gl}")
    print(f"nivel de significancia:   {alpha}")

    if ji_calc < valor_critico:
        print("\nResultado: No se rechaza la hipótesis nula (H0).")
        print("Los números se distribuyen uniformemente.")
    else:
        print("\nResultado: Se rechaza la hipótesis nula (H0).")
        print("Los números NO se distribuyen uniformemente.")


prueba_frecuencias(datos)
```
